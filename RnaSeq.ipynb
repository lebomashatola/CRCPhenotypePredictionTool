```python
import os
import pandas as pd
```

```python
def read_files_create_dataframe(path):
    
    df = pd.DataFrame()
    directory = '/Users/lebohangmashatola/downloads/' + path + '/C/'
    count = 0
    colnames = []
    
    for file in os.listdir(directory):
                
        filename = os.fsdecode(file)
        read_data = directory + '/' + filename
                        
        try:
            files = os.listdir(read_data)
                                    
        except:
            continue
        
        for i in files:
            
            if i.endswith(".tsv"):
                                
                df_files = read_data + '/' + i
                df_ = pd.read_csv(df_files, sep='\t', header=1)
                
                df_ = df_.iloc[4:] #First 4 rows descriptions of columns and should be removed 
            
                df_ = df_[df_['gene_type'].str.match('protein_coding')]
                df_ = df_.set_index('gene_name')
                df_ = df_[['fpkm_uq_unstranded']] #Other options include tpm unstranded; fpkm_unstranded 
                        
                df = pd.concat([df, df_], axis=1)
                
                count += 1 
                colnames.append('P' + str(count))
                
                
    
    df = df.T
    df = pd.DataFrame(df)
    df.index = colnames 
    
    output_directory = '/Users/lebohangmashatola/Downloads/CRCPhenotypePredictionTool-main/Expression/' + path + '.csv'
    df.to_csv(output_directory)
    
    return df
```
